```python
import re

# Abrir o arquivo
with open('../ALL_montage-0-50d-filelist.txt', 'r') as file:
    lines = file.readlines()

# Lista para armazenar os dados
file_sizes = []

# Expressão regular para extrair o tamanho do arquivo e o nome do arquivo
pattern = r'-rw-r--r-- 1 root root[ ]{1,}(\d+) \w+[ ]{1,}\d+ \d+:\d+ (\S+)'

# Iterar sobre as linhas do arquivo
for line in lines:
    match = re.match(pattern, line)
    if match:
        file_name = match.group(2)
        file_size_bytes = int(match.group(1))
        file_size_mb = file_size_bytes / (1000 * 1000)  # Convertendo bytes para MB
        file_sizes.append((file_name, file_size_mb))
#    else:
#        print('Match not happen')

# Ordenar os dados pelo tamanho do arquivo
sorted_file_sizes = sorted(set(file_sizes), key=lambda x: x[0])

# Obter o tamanho máximo do nome do arquivo
max_file_name_length = max(len(file_name) for file_name, _ in sorted_file_sizes)

# Imprimir os resultados alinhando os tamanhos dos arquivos
for file_name, file_size in sorted_file_sizes:
    print(f'{file_name.ljust(max_file_name_length)} \t {file_size:.6f}')
    
# Imprimir os resultados
# for file_name, file_size in file_sizes:
#     print(f'{file_name} \t {file_size}')
```

```python
# Lista para armazenar os dados para ordenação
data = []

# Abrir o arquivo para leitura
with open('../wf-montage-050d_job_durations.txt', 'r') as file:
    # Ler as linhas do arquivo
    lines = file.readlines()

    # Encontrar o maior tamanho de nome de atividade
    max_activity_name_length = max(len(line.split()[0]) for line in lines)

    # Iterar sobre as linhas do arquivo
    for line in lines:
        # Dividir a linha em colunas separadas por espaço em branco
        columns = line.split()

        # Se a linha não tiver informações, continuar para a próxima linha
        if len(columns) == 0:
            continue

        # Remover as colunas 'COMPLETIONS' (índice 1) e 'AGE' (última coluna)
        del columns[1]  # 'COMPLETIONS'
        del columns[-1]  # 'AGE'
        
        # Obter o nome da atividade e a duração
        activity_name = columns[0]
        duration = columns[1]

        # Remover o 's' de segundo, se presente
        duration = duration.replace('s', '')

        # Verificar se os últimos 7 caracteres são números
        if activity_name[-7:].isdigit():
            pass
            # print("Os últimos 7 caracteres são números.")
        else:
            continue
        
        # Se a duração contiver 'm' de minutos, converter para segundos
        if 'm' in duration:
            # Remover 'm' e converter para segundos
            minutes = int(duration.replace('m', ''))
            seconds = minutes * 60
            duration = str(seconds)

        activity_id = int(activity_name[-7:])
        
        # Alinhar o nome da atividade
        activity_name = columns[0].ljust(max_activity_name_length)
        
        # Atualizar a coluna de duração na linha
        columns[0] = activity_name
        # columns[1] = duration_aligned

        # Adicionar dados à lista para ordenação
        data.append((activity_id, activity_name, duration))

# Ordenar os dados pelo número final do nome do arquivo
data.sort(key=lambda x: x[0])

# Imprimir os dados com a nova coluna ID
print("NAME", "\t\t\t", "DURATION")
for item in data:
    print(item[1], "\t", item[2])
    
        # # Alinhar a coluna de duração
        # duration_aligned = '{:<7}'.format(duration)

        # # Alinhar o nome da atividade
        # activity_name = columns[0].ljust(max_activity_name_length)

        # # Atualizar a coluna de duração na linha
        # columns[0] = activity_name
        # columns[1] = duration_aligned

        # # Imprimir a linha atualizada
        # print(' \t'.join(columns))
```
